```python
df = pd.read_csv("E:\\Internship\\Used_Bikes.csv")
df.shape

df.drop_duplicates(inplace=True)
df.shape

brand_name = list(df['brand'].value_counts().keys())
total_bikes = list(df['brand'].value_counts().values)
plt.figure(figsize=(10,4))
plt.bar(brand_name,total_bikes,color='orange')
plt.xticks(rotation=80)
plt.show()
```

```python
bike_type = list(df['owner'].value_counts().keys())
total_bikes = list(df['owner'].value_counts().values)
plt.figure(figsize=(10,4))
plt.bar(bike_type,total_bikes,color='orange')
plt.xticks(rotation=80)
plt.show()
```
